```python
sns.boxplot(rf['math_score'])
Q3 = rf.math_score.quantile(.75)
Q1 = rf.math_score.quantile(.25)
IQR = Q3 - Q1
Median = rf.math_score.median()
print("Q1 Value:",Q1)
print("Median Value:",rf.math_score.median())
print("Q3 Value:",Q3)
print("Upper whisker limit:",(Q3 + 1.5*IQR))
print("Lowerr whisker limit:",(Q1 - 1.5*IQR))
# the median math marks out of all marks is 66
```

```python
sns.boxplot(rf['writing_score'])
Q3 = rf.writing_score.quantile(.75)
Q1 = rf.writing_score.quantile(.25)
IQR = Q3 - Q1
Median = rf.writing_score.median()
print("Q1 Value:",Q1)
print("Median Value:",rf.writing_score.median())
print("Q3 Value:",Q3)
print("Upper whisker limit:",(Q3 + 1.5*IQR))
print("Lowerr whisker limit:",(Q1 - 1.5*IQR))
# the median writing marks out of all marks is 69
```

```python
sns.boxplot(rf['reading_score'])
Q3 = rf.reading_score.quantile(.75)
Q1 = rf.reading_score.quantile(.25)
IQR = Q3 - Q1
Median = rf.reading_score.median()
print("Q1 Value:",Q1)
print("Median Value:",rf.reading_score.median())
print("Q3 Value:",Q3)
print("Upper whisker limit:",(Q3 + 1.5*IQR))
print("Lowerr whisker limit:",(Q1 - 1.5*IQR))
# the median reading marks out of all marks is 70
```
